```python
import pandas as pd
```

```python
PG = pd.read_csv("Power generation.csv")
PG['date'] = pd.to_datetime(PG['Date'], format='%Y%m')
PG.drop(['Date','單位'], axis=1, inplace=True)
PG['year'] = PG['date'].dt.year
PG['month'] = PG['date'].dt.month
PG_filtered = PG[PG['date'] != '2024-01-01']

PG_filtered
```

```python
columns = PG_filtered.columns
columns = columns[:-3]
columns
```

```python
PG_year = PG_filtered.groupby('year')[columns].sum().reset_index()
PG_year
```

```python
PG_filtered.to_csv('powerGenerationM.csv', sep=',', index=False, encoding='utf-8')
PG_year.to_csv('powerGenerationY.csv', sep=',', index=False, encoding='utf-8')
```

```python
# PG_filtered

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
PG_2016 = PG_filtered[PG_filtered['date'] >= '2016-01-01']
PG_2016 = PG_2016.copy()
PG_2016['X'] = (PG_2016['year'] - 2016) * 12 + PG_2016['month']
PG_2016
```

```python
X_train = PG_2016[['X']]
y_train = PG_2016['NPGRenewableTotal']/PG_2016['NPGOverall']*100
X_train
```

```python
import pandas as pd

# Create a date range from January 2024 to December 2025 with monthly frequency
date_range = pd.date_range(start='2016-01-01', end='2025-12-01', freq='MS')

# Create a DataFrame with the date range
df = pd.DataFrame({'date': date_range})
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
df['X'] = (df['year'] - 2016) * 12 + df['month']

# Display the DataFrame
X_test = df[['X']]
X_test
```

```python
linear_model = LinearRegression()
polynomial_model = LinearRegression()

# 4. Train the Models
linear_model.fit(X_train, y_train)

# Transform features for polynomial regression
poly_features = PolynomialFeatures(degree=2)  # Degree 2 for square term
X_train_poly = poly_features.fit_transform(X_train)
polynomial_model.fit(X_train_poly, y_train)

# 5. Evaluate the Models
linear_predictions = linear_model.predict(X_test)

# Transform features for polynomial regression
X_test_poly = poly_features.transform(X_test)
polynomial_predictions = polynomial_model.predict(X_test_poly)


# 6. Visualize the Results
plt.plot(df['date'], linear_predictions, color='blue', label='Linear Regression')
plt.plot(df['date'], polynomial_predictions, color='red', label='Polynomial Regression')
plt.xlabel('Time')
plt.ylabel('Power Generation')
plt.title('Renewable Power Generation Prediction')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.utils import resample

# Assume X_train, X_test, y_train, and y_test are already defined

# 1. Initialize models
linear_model = LinearRegression()
polynomial_model = LinearRegression()

# 2. Train the models
linear_model.fit(X_train, y_train)

# Transform features for polynomial regression
poly_features = PolynomialFeatures(degree=2)  # Degree 2 for square term
X_train_poly = poly_features.fit_transform(X_train)
polynomial_model.fit(X_train_poly, y_train)

# 3. Predictions
linear_predictions = linear_model.predict(X_test)

# Transform features for polynomial regression
X_test_poly = poly_features.transform(X_test)
polynomial_predictions = polynomial_model.predict(X_test_poly)

# 4. Confidence Intervals
# Bootstrap resampling for confidence intervals
n_bootstraps = 1000
bootstrap_linear_predictions = np.zeros((X_test.shape[0], n_bootstraps))
bootstrap_polynomial_predictions = np.zeros((X_test.shape[0], n_bootstraps))

for i in range(n_bootstraps):
    # Resample with replacement
    X_boot, y_boot = resample(X_train, y_train)
    linear_model.fit(X_boot, y_boot)
    bootstrap_linear_predictions[:, i] = linear_model.predict(X_test)

    X_boot_poly, y_boot_poly = resample(X_train_poly, y_train)
    polynomial_model.fit(X_boot_poly, y_boot_poly)
    bootstrap_polynomial_predictions[:, i] = polynomial_model.predict(X_test_poly)

# Calculate 95% confidence intervals
ci_lower_linear = np.percentile(bootstrap_linear_predictions, 2.5, axis=1)
ci_upper_linear = np.percentile(bootstrap_linear_predictions, 97.5, axis=1)

ci_lower_poly = np.percentile(bootstrap_polynomial_predictions, 2.5, axis=1)
ci_upper_poly = np.percentile(bootstrap_polynomial_predictions, 97.5, axis=1)

# 5. Visualize the Results
plt.plot(df['date'], linear_predictions, color='blue', label='Linear Regression')
plt.plot(df['date'], polynomial_predictions, color='red', label='Polynomial Regression')

plt.fill_between(df['date'], ci_lower_linear, ci_upper_linear, color='lightblue', alpha=0.5, label='Linear Regression 95% CI')
plt.fill_between(df['date'], ci_lower_poly, ci_upper_poly, color='lightcoral', alpha=0.5, label='Polynomial Regression 95% CI')

plt.xlabel('Time')
plt.ylabel('Power Generation')
plt.title('Renewable Power Generation Prediction with 95% CI')
plt.legend()
plt.show()
```

```python
import pandas as pd

# Create a DataFrame to store the results
results_df = pd.DataFrame({
    'date': df['date'],
    'LinearRegressionPrediction': linear_predictions,
    'LinearRegressionCILower': ci_lower_linear,
    'LinearRegressionCIUpper': ci_upper_linear,
    'PolynomialRegressionPrediction': polynomial_predictions,
    'PolynomialRegressionCILower': ci_lower_poly,
    'PolynomialRegressionCIUpper': ci_upper_poly
})

# Output the DataFrame to a CSV file
results_df.to_csv('predictions_with_confidence_intervals.csv', index=False)
```
